- bookID: Unique identification number fro each book
- title: Name under which book was published
- authors: Name of the Authors of the book
- average_rating: Avarage rating of the book recevied in total.
- isbn: International standarded book number
- isbn13: 13 digit isbn to identify the book
- language_code: Primary Language of the book
- num_pages: Number of pages the book containes
- ratings_count: Total Number of ratings the book recevied.
- text_reviews_count: Total number of written reviews recevied.
- publication_date: Date when the book was first published
- publisher: Name of the Pulishers

# Data Preprocessing

```python
df.head(2)
```

```python
def num_to_obj(x):
    if x >0 and x <=1:
        return "between 0 and 1"
    if x > 1 and x <= 2:
        return "between 1 and 2"
    if x > 2 and x <=3:
        return "between 2 and 3"
    if x >3 and x<=4:
        return "between 3 and 4"
    if x >4 and x<=5:
        return "between 4 and 5"
df['rating_obj'] = df['average_rating'].apply(num_to_obj)
```

```python
df['rating_obj'].value_counts()
```

```python
rating_df = pd.get_dummies(df['rating_obj'])
rating_df.head()
```

```python
df.columns
```

```python
language_df = pd.get_dummies(df['language_code'])
language_df.head()
```

```python
features = pd.concat([rating_df,language_df, df['average_rating'],
                    df['ratings_count'], df['title']], axis = 1)
features.set_index('title', inplace= True)
features.head()
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler()
features_scaled = scaler.fit_transform(features)
```

```python
@interact
def BookRecomender(book_name = list(df['title'].value_counts().index)):
    book_list_name = []
    book_id = df[df['title'] == book_name].index
    book_id = book_id[0]
    for newid in idlist[book_id]:
        book_list_name.append(df.iloc[newid].title)
    return book_list_name
```

```python
import tkinter as tk
from tkinter import ttk, messagebox, font
import pandas as pd
from ttkwidgets.autocomplete import AutocompleteCombobox

df.drop_duplicates()
# Assuming df and idlist are defined elsewhere in your code
# df = ...
# idlist = ...

def BookRecommender(book_name):
    book_list_details = []
    try:
        book_id = df[df['title'] == book_name].index
        if not book_id.empty:
            book_id = book_id[0]
            for newid in idlist[book_id]:
                book_details = df.iloc[newid]
                details = f"Author: {book_details['authors']} - Rating: {book_details['average_rating']} - Publication Date: {book_details['publication_date']}"
                book_list_details.append((book_details['title'], details))
        else:
            messagebox.showinfo("Info", "Book not found in the dataframe.")
    except Exception as e:
        messagebox.showerror("Error", str(e))
    
    return book_list_details

def on_select(event):
    # Get the selected item from the listbox
    widget = event.widget
    selection = widget.curselection()
    if selection:
        selected_item = widget.get(selection)
        
        # If the selected item is a book, show its details
        if selected_item in initial_books:
            show_book_details(selected_item)

def search_book():
    # Get the input from the entry widget
    book_name = entry.get()
    
    # Get recommendations for the searched book
    recommendations = BookRecommender(book_name)
    
    # Clear previous suggestions and add new ones
    suggestion_listbox.delete(0, tk.END)
    for book, details in recommendations:
        suggestion_listbox.insert(tk.END, book)

def show_book_details(book_name):
    try:
        # Get the details of the selected book from the dataframe
        book_details = df[df['title'] == book_name].iloc[0]
        
        # Create a new window to display book details
        book_window = tk.Toplevel(root)
        book_window.title(book_name)
        
        # Display book details in labels
        details_labels = []
        for i, (label, value) in enumerate(book_details.iteritems()):
            label_text = f"{label}:"
            value_text = value
            
            label_widget = tk.Label(book_window, text=label_text, font=("Playfair Display", 12, "bold"))
            label_widget.grid(row=i, column=0, sticky="w", padx=10, pady=5)
            value_widget = tk.Label(book_window, text=value_text, font=("Playfair Display", 12))
            value_widget.grid(row=i, column=1, sticky="w", padx=10, pady=5)
            
            details_labels.append((label_widget, value_widget))
    except IndexError:
        messagebox.showinfo("Info", "Book details not found.")

def go_back():
    # Clear the search bar
    entry.delete(0, tk.END)
    
    # Clear previous suggestions and repopulate the listbox with initial books
    suggestion_listbox.delete(0, tk.END)
    for book, details in initial_books_details:
        suggestion_listbox.insert(tk.END, book)

# Create the main window
root = tk.Tk()
root.title("Book Recommender")
root.geometry("1200x600")  # Set window size

# Register Playfair Display font
font.families()

# Create a frame to contain the search bar and search button
search_frame = tk.Frame(root)
search_frame.pack(pady=20, padx=20, fill=tk.X)

search_label = tk.Label(search_frame, text="Enter book name for suggestion", font=("Playfair Display", 12), anchor='w')
search_label.pack(side=tk.TOP, pady=(0, 5), padx=5, fill=tk.X)

# Create the entry widget for input
entry_placeholder = "Search here"
entry = AutocompleteCombobox(search_frame, font=("Playfair Display", 12))
entry.set_completion_list(df['title'].tolist())
entry.insert(0, entry_placeholder)
entry.pack(side=tk.LEFT, padx=5, ipady=5, fill=tk.X, expand=True)

def on_entry_focus_in(event):
    if entry.get() == entry_placeholder:
        entry.delete(0, tk.END)

def on_entry_focus_out(event):
    if entry.get() == "":
        entry.insert(0, entry_placeholder)

entry.bind("<FocusIn>", on_entry_focus_in)
entry.bind("<FocusOut>", on_entry_focus_out)

def search_button_clicked():
    search_book()

# Create the search button
search_button = ttk.Button(search_frame, text="Search", command=search_button_clicked, style='Search.TButton')
search_button.pack(side=tk.LEFT, padx=5)

# Configure style for the search button
style = ttk.Style()
style.configure('Search.TButton', font=('Playfair Display', 12), padding=10)

# Create a frame to contain the listbox
frame = tk.Frame(root)
frame.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)

# Create the listbox for suggestions
suggestion_listbox = tk.Listbox(frame, font=("Playfair Display", 12), selectmode=tk.SINGLE)
suggestion_listbox.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

# Bind the on_select function to listbox selection event
suggestion_listbox.bind("<Double-Button-1>", on_select)

# Populate the listbox with initial book titles
initial_books_details = []
initial_books = df['title'].tolist()
for book in initial_books:
    details = BookRecommender(book)
    if details:
        initial_books_details.append((book, details[0]))
        suggestion_listbox.insert(tk.END, f"{book}\n{' ' * 30}")  # Add some space between book titles

# Create the "Go Back" button
go_back_button = ttk.Button(root, text="Go Back", command=go_back, style='GoBack.TButton')
go_back_button.pack(pady=(10, 20), padx=20, ipady=5, fill=tk.X)

# Configure style for the "Go Back" button
style.configure('GoBack.TButton', foreground='black', background='black', font=('Playfair Display', 12), padding=10)
style.map('GoBack.TButton', background=[('active', 'gray')], foreground=[('active', 'white')])
root.mainloop()
```
